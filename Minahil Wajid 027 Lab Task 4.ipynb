```python
#Question no 3:
def calculate_discount(price, discount=15):
    discount_amount = price * (discount / 100)
    return discount_amount
calculate_discount(2000, 15)
print(calculate_discount(2000, 15))
```

```python
#Question no 5:
def find_max(*numbers):
    return max(numbers)
print(find_max(2, 5, 8, 1))
```

```python
#Question no 6:
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)
print(factorial(5))
```

```python
#Question no 7:
def square(x):
    return x * x
def apply_function(function, numbers):
    result = []
    for num in numbers:
        result.append(function(num))
    return result
print(apply_function(square, [1, 2, 3, 4]))
```

```python
#Question no 9:
def square(x):
    return x * x
def apply_operation(function, numbers):
    result = []
    for num in numbers:
        result.append(function(num))
    return result
print(apply_operation(square, [1, 2, 3]))
```

```python
#Question no 10:
import time
def timer(function):
    def wrapper():
        start = time.time()
        function()
        end = time.time()
        print("Execution Time:", end - start)
    return wrapper
@timer
def test():
    for i in range(100000):
        pass
test()
```
