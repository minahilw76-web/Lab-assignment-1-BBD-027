```python
#Question no 10:
secret = 25
guess = int(input("Guess number: "))
count = 1
while guess != secret:
    if guess > secret:
        print("Too high")
    else:
        print("Too low")    
    guess = int(input("Try again: "))
    count = count + 1
print("Correct in", count, "tries")
```
