```python
#Question no 4:
orders = [
    {"product": "cargo_pants", "category": "clothing"},
    {"product": "saari", "category": "clothing"},
    {"product": "laptop", "category": "electronics"},
    {"product": "phone", "category": "electronics"},
    {"product": "shirt", "category": "clothing"}]
category_count = {}
for order in orders:
    cat = order["category"]
    category_count[cat] = category_count.get(cat, 0) + 1
most_popular = max(category_count, key=category_count.get)
print("Most popular category:", most_popular)
```

```python
#Question no 6:
inventory = {}
def add_product(name, quantity, price):
    inventory[name] = {"quantity": quantity, "price": price}
def update_quantity(name, quantity):
    if name in inventory:
        inventory[name]["quantity"] += quantity
def total_value():
    total = 0
    for item in inventory.values():
        total += item["quantity"] * item["price"]
    return total
def low_stock(threshold=5):
    return {k: v for k, v in inventory.items() if v["quantity"] < threshold}
add_product("pen", 10, 20)
add_product("notebook", 3, 100)
update_quantity("pen", 5)
print("Total Value:", total_value())
print("Low Stock:", low_stock())
```

```python
#Question no 7:
students = {
    "Mishaal": {"accounting": 80, "psychology": 75},
    "Anaya": {"accounting": 90, "psychology": 85},
    "Hamza": {"accounting": 70, "psychology": 65}}
averages = {}
for name, subjects in students.items():
    avg = sum(subjects.values()) / len(subjects)
    averages[name] = avg
print("Averages:", averages)
accounting_scores = {name: sub["accounting"] for name, sub in students.items()}
highest = max(accounting_scores, key=accounting_scores.get)
lowest = min(accounting_scores, key=accounting_scores.get)
print("Highest in accounting:", highest)
print("Lowest in accounting:", lowest)
sorted_students = sorted(averages.items(), key=lambda x: x[1], reverse=True)
print("Sorted:", sorted_students)
```

```python
#Question no 8:
# H
products = {
    "shirt": {"price": 1000, "quantity": 10},
    "jeans": {"price": 2000, "quantity": 5}
}

cart = {}

def add_to_cart(item, qty):
    if item in products and products[item]["quantity"] >= qty:
        cart[item] = cart.get(item, 0) + qty
        products[item]["quantity"] -= qty

def remove_from_cart(item, qty):
    if item in cart:
        cart[item] -= qty
        if cart[item] <= 0:
            del cart[item]
def calculate_total(discount=0, tax=0):
    total = 0
    for item, qty in cart.items():
        total += products[item]["price"] * qty    
    total -= total * discount
    total += total * tax
    return total
add_to_cart("shirt", 2)
add_to_cart("jeans", 1)
print("Cart:", cart)
print("Total:", calculate_total(discount=0.1, tax=0.05))
```
